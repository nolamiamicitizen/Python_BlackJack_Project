## Card Class
###### Card class is created; this game only uses one deck that gets shuffled after each turn.  Suits and Faces are defined along with card values.

```python
class Card():

    
    def __init__(self, faceNum,suitNum,ACE_case=False):
        self.faceNum = faceNum
        self.suitNum = suitNum
        self.ACE_case = ACE_case
        
    
    def getCardName(self):
        nameSuit = ['Spades', 'Hearts', 'Clubs', 'Diamonds']
        nameFace = ['Ace', '2','3','4','5','6','7','8','9','10', 'Jack', 'Queen',  'King']
        
        return "%s of %s" % (nameFace[self.faceNum], nameSuit[self.suitNum])

    def card_value(self):
        if self.faceNum == 0:
            self.ACE_case = True
            return (1,11)
        elif 1 <= self.faceNum <= 9:
            return self.faceNum+1
        else:
            return 10
    
def MakeDeck():
    deck = []
    for suitNum in range(4):
        for faceNum in range(13):
            newCard = Card (faceNum, suitNum)
            deck.append(newCard)
    return deck

deck = MakeDeck()
for card in deck:
    print card.getCardName()
    #print card.faceNum
    
class B:
    def __init__(self):
        pass
```

## Dealer Class
##### Dealer class is created with less attributes than a player and a default movement for acting in the game (Dealer Turn)

```python
class Dealer(object):
    
    def __init__(self,CARD_amount=0,BustedStatus=False,ACES_held=0,blackjackstatus=0):
            self.ACES_held = ACES_held
            self.CARD_amount = CARD_amount
            self.BustedStatus = BustedStatus
            self.blackjackstatus = blackjackstatus
            
            #main card list is created, to be a unique list for each player with these cards
            card_list = []
            for object_name in deck:
                card_list.append(object_name)
                self.card_list = card_list
            
            #determines busted status
    def BustedCheck(self):
        if self.CARD_amount > 21:
            self.BustedStatus = True
            
        else:
            self.BustedStatus = False
            
            
    def add_CARDamount(self,CARD_points):
        self.CARD_amount += CARD_points
        
    def reset_CARDamount(self):
        self.CARD_amount = 0
    
    def reset_CardsHeld(self):
        del self.card_list[:]
        

    def CalculateCARD_amount(self):
        #sets up for calculations
        self.CARD_amount = 0
        self.ACES_held = 0
        beforeAcesCardAmount = 0
        #tracks the number of Aces and adds up other card values
        for card in self.card_list:
            if card.faceNum == 0:
                self.ACES_held += 1
                continue
            else:
                beforeAcesCardAmount += card.card_value()
                continue
        
        AcesHighestValue = self.ACES_held * 11
        self.CARD_amount = beforeAcesCardAmount + AcesHighestValue
        
        #calculates the hig = hest amount based on the value of the Aces that won't go over 21, if possible
        while self.ACES_held <> 0:
            #if over, subtracts the additional 10 points. of card value of the 11 point Ace, counts down the ace
            if self.CARD_amount > 21:
                self.CARD_amount -= 10
                self.ACES_held -= 1
                #continue
            else:
                #If under, just counts down the ace
                self.ACES_held -= 1
                #continue
        for card in self.card_list:
            if card.faceNum == 0:
                self.ACES_held += 1
                continue
                
    def DEALERdisplayopenCard(self):
            print self.card_list[0].getCardName()
            
    def listouttheCards(self):
        for card in self.card_list:
            print card.getCardName()
    
    #Dealer Turn is an embedded function
    def Dealer_Turn(self):
        global Card_deck_position
        self.CalculateCARD_amount()
        self.BustedCheck()
    
    #dealer draws until at or over 17
    #soft 17 requires a draw
        while self.CARD_amount < 17:
            self.card_list.append(deck[Card_deck_position])
            Card_deck_position += 1
            self.CalculateCARD_amount()
   
        self.BustedCheck()
```
